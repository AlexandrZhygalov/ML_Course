One hot Encoding (трансформація інформації у true-false формат)

Використовуємо датасет adult

header - параметр, що замінює назви колон на цифри

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
sns.set_theme()
```

```python
X = pd.read_csv('Datasets/adult.data', header=None)
X.shape
```

```python
X.head()
```

Проводимо one hot encoding для колонок (матриця розширяється)

```python
X2 = pd.get_dummies(X, columns=[1,3])
X2.head()
```

Shape матриці змінюється з (32561, 15) на (32561, 38)

```python
X2.shape
```

iloc - пошук по індексу. loc - пошук по назві

Виведення тільки колонки, чи зарплата більше/менше 50000

```python
y = X2.loc[:,14]
```

```python
y
```

Виведення, які аргументи є в цій матриці

```python
vals = np.unique(y)
```

```python
vals
```

Шукаємо, у кого зарплата менше 50000

```python
idx = X2.loc[:,14]==vals[0]
```

```python
idx
```

Трансформуємо дату таким чином, що замість <=50k та >50k буде 1 або 0 відповідно

```python
X2.loc[idx,14] = 0
X2.loc[~idx,14] = 1
```

```python
X2.loc[:,14]
```

Виведення зміненої датабази

```python
np.unique(X2.loc[:,14])
```

```python
X2
```

Збереження усіх змін у файл (тут згадаємо індуса, що поклав бд заради інфи про зарплати)

```python
X2.to_csv("Datasets/trash_data.csv")
```
